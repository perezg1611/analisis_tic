# Análisis del Sector TIC en Colombia
## Basado en: Boletín Trimestral MinTIC | 2023 – 2025

**Autor:** Jhonatan Smith Pérez Guerrero  
**Fecha:** Julio 2026  
**Fuente de datos:** Ministerio de Tecnologías de la Información 
y las Comunicaciones — colombiatic.mintic.gov.co  
**Herramientas:** Python · Pandas · Matplotlib · Seaborn · Power BI · SQL

## Contexto y Objetivo

El presente análisis explora el comportamiento del sector de 
Telecomunicaciones en Colombia a partir del Boletín Trimestral 
publicado por el MinTIC, que consolida información reportada 
por los proveedores de redes y servicios del país.

El análisis cubre desde el periodo **2023 Q2 hasta 2025 Q4** e integra 
datos de abonados móviles, tráfico de datos, cobertura municipal 
y resultados financieros de los operadores.

Nota: Abonados entíendase como suscriptor de un contrato con el operador.

### Preguntas que guiaron el análisis

1. ¿Qué operador lidera el mercado de abonados móviles y cómo se 
   distribuyen por tecnología y segmento?
2. ¿Cómo evolucionó el tráfico de datos móviles por operador 
   entre 2023 y 2025?
3. ¿Qué departamentos lideran la adopción del 5G en Colombia?
4. ¿Existe relación entre el número de líneas en servicio y los 
   ingresos operacionales por operador?

## Hallazgo 1 — Distribución de Abonados Móviles por Tecnología

![abonados](../dashboard/capturas/abonados_por_tecnologia_2025_Q4.png)

El mercado móvil colombiano es ampliamente dominado por el segmento 
**prepago**, con más de 25 millones de abonados frente a 
aproximadamente 1.900 en pospago. Esto refleja las condiciones 
socioeconómicas del país, donde los planes de contrato mensual 
tienen menor penetración que el acceso por recarga.

**Claro (COMCEL)** lidera en ambos segmentos con más del 60% 
del mercado prepago. En tecnología, el **4G domina** la 
distribución de abonados, mientras que el 5G comienza a aparecer 
en los operadores más grandes como una capa incipiente.

> **Nota metodológica:** La diferencia entre abonados (internet móvil) 
y líneas en servicio (voz + datos) explica por qué el total de 
líneas supera los 90 millones en un país de 52 millones de habitantes. 
Una persona puede tener múltiples SIMs activas simultáneamente.

## Hallazgo 2 — Evolución del Tráfico de Datos 2023–2025

![trafico](../dashboard/capturas/trafico_por_operador_2023_2025.png)

El tráfico de datos móviles ha crecido de forma sostenida durante 
el periodo analizado. **Claro** mantiene un volumen consistentemente 
superior a los 250.000 TB por trimestre, alcanzando aproximadamente 
**370.000 TB en Q4 2025**.

Un hallazgo relevante es la **convergencia entre Tigo y Movistar** 
hacia finales de 2025 — Movistar muestra una tendencia de crecimiento 
más pronunciada, acercándose por primera vez a los niveles de tráfico 
de Tigo. Esto sugiere una ganancia de participación de mercado por 
parte de Movistar en el segmento de datos.

## Hallazgo 3 — Cobertura 5G por Departamento

![5g](../dashboard/capturas/cobertura_5G_por_departamento.png)

Al analizar los datos de cobertura municipal se encontró que el 
**100% de los municipios reportados tienen cobertura 4G** de al 
menos un operador, lo que indica una penetración consolidada de 
esta tecnología en el territorio nacional.

La brecha digital actual en Colombia **ya no es de 4G vs tecnologías 
anteriores**, sino de **5G vs sin 5G**. Antioquia lidera con 27 
municipios con cobertura 5G, seguido de Cundinamarca con 15 y 
Valle del Cauca con 11.

Los departamentos de frontera y selva como **Guainía, Vaupés y 
Vichada** no registran ningún municipio con cobertura 5G, 
evidenciando que la brecha digital geográfica persiste en las 
zonas más remotas del país.

> **Nota metodológica:** La columna LTE aparece vacía en el dataset, 
posiblemente porque el MinTIC la consolida dentro de la categoría 4G.

## Hallazgo 4 — Relación entre Líneas en Servicio e Ingresos

![ingresos](../dashboard/capturas/lineas_vs_ingresos.png)

Existe una **correlación positiva clara** entre el número de líneas 
en servicio y los ingresos operacionales. A mayor volumen de 
suscriptores, mayores ingresos — lo que confirma que el volumen 
de líneas es el principal driver financiero del sector.

El caso más interesante es **Movistar**, que con 23.1M de líneas 
genera 146B en ingresos, superando proporcionalmente a **Tigo** 
que con 17.5M de líneas genera 104B. Esto sugiere que Movistar 
tiene un **ingreso promedio por usuario (ARPU) más alto**, 
posiblemente por mayor proporción de clientes pospago o 
planes de mayor valor.

Los Operadores Móviles Virtuales (**OMVs**) como Almacenes Éxito 
y Virgin Mobile, con 2.4M y 3.3M de líneas respectivamente, 
generan ingresos mínimos al no tener infraestructura propia y 
revender la red de los operadores grandes.

## Conclusión General

El sector TIC móvil en Colombia muestra un mercado **consolidado 
y en crecimiento**, con Claro como líder indiscutible en todas 
las dimensiones analizadas — abonados, tráfico e ingresos. 

Sin embargo, el análisis revela dinámicas competitivas interesantes: 
Movistar demuestra mayor eficiencia en generación de ingresos por 
usuario y un crecimiento acelerado en tráfico de datos hacia 2025, 
mientras que Tigo mantiene una base amplia de usuarios con menor 
rentabilidad por línea.

La agenda pendiente del sector es clara: **la expansión del 5G** 
sigue concentrada en los departamentos más desarrollados del país, 
dejando a las regiones de frontera y selva sin acceso a la 
tecnología de próxima generación. Cerrar esta brecha será el 
principal desafío del sector en los próximos años.

## Limitaciones y Trabajo Futuro

### Limitaciones del análisis
- El dataset solo incluye municipios donde los operadores 
  tienen presencia registrada ante el MinTIC
- La columna LTE aparece vacía — posiblemente consolidada en 4G 
  por los operadores al momento del reporte
- Los datos de pospago muestran volúmenes significativamente 
  menores que prepago, lo que puede reflejar diferencias en 
  la metodología de reporte

### Análisis futuros propuestos
- Brecha digital en acceso fijo a internet por departamento
- Cálculo del ARPU (ingreso promedio por usuario) por operador 
  y su evolución trimestral
- Correlación entre cobertura 5G y nivel de desarrollo 
  socioeconómico por departamento
- Comparación de la penetración móvil colombiana con otros 
  países de América Latina
