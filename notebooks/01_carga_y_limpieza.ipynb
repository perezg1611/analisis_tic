```python
import pandas as pd
```

```python
#Leemos excel e identificamos los nombres de las pestañas
path = r'C:/Users/perez/Documents/Repositorios/analisis_tic/data/raw/data_raw.xlsx'

xl = pd.ExcelFile(path)
print(xl.sheet_names)
```

##### Las pestañas que nos interesan para responder las preguntas son: Tablas: 2, 6, 8, 15, 12 y 14

```python
#Pestañas de interés
df_tabla2 = pd.read_excel(path, sheet_name="2", skiprows=5)
df_tabla6 = pd.read_excel(path, sheet_name="6", skiprows=5)
df_tabla8 = pd.read_excel(path, sheet_name="8", skiprows=5)
df_tabla15 = pd.read_excel(path, sheet_name="15", skiprows=5)
df_tabla12 = pd.read_excel(path, sheet_name="12", skiprows=5)
df_tabla14 = pd.read_excel(path, sheet_name="14", skiprows=5)
```

```python
#Validación de los dataframes y  valores nulos
print(df_tabla14.head())
print(df_tabla14.shape)
print(df_tabla14.isnull().sum())
```

##### Datos separados y convertidos a CSV para mejor procesamiento

```python
df_tabla2.to_csv(f"../data/processed/accesos_fijos.csv", index=False, encoding="utf-8-sig")
df_tabla6.to_csv(f"../data/processed/abonados_moviles.csv", index=False, encoding="utf-8-sig")
df_tabla8.to_csv(f"../data/processed/trafico_movil.csv", index=False, encoding="utf-8-sig")
df_tabla12.to_csv(f"../data/processed/lineas_moviles.csv", index=False, encoding="utf-8-sig")
df_tabla14.to_csv(f"../data/processed/ingresos_por_voz.csv", index=False, encoding="utf-8-sig")

#Optimizamos el nombre de las columnas de este último dataframe
df_tabla15 = df_tabla15.rename(columns={
    "COBERTURA 2G":   "2G",
    "COBERTURA 3G":   "3G",
    "COBERTURA HSPA+, HSPA+DC":"HSPA+",
    "COBERTURA_4G":   "4G",
    "COBERTURA_LTE":  "LTE",
    "COBERTURA_5G":   "5G"
})

df_tabla15.to_csv(f"../data/processed/cobertura_movil_por_municipio.csv", index=False, encoding="utf-8-sig")
```
